```python
a = 2
b = "2"
print(" a = ", a, "type = ", type(a))
print(" b = ", b, "type = ", type(b))
```

```python
print ("Execution Started")
print ("converting a into string and then doing + operation, a + b = ", str(a) + b)
print ("converting b into int and then doing + operation, a + b = ", a + int(b))
print ("Execution Ended")
```
